```python
import glob 
import re #regex
import os
import jams
import librosa
import muda
```

```python
path_database = '/home/deborah/Área de Trabalho/TIA_tutorials/urbansound/UrbanSound8K/audio/train/'
path_backgound_noise = '/home/deborah/Área de Trabalho/TIA_tutorials/urbansound/UrbanSound8K/background_noise/'
```

```python
def jams_generation(audio_name):
    
    #propriedades do jams
    data_source='UrbanSound8K'
    name='Deborah'
    email='deborah.vm@ufpi.edu.br'
    
    #Capturando a duracao do sinal
    y, sr = librosa.load(audio_name, sr=None)
    duration = librosa.get_duration(y=y, sr=sr)
            
    #Gerando o arquivo de anotacao jam 
    jam = jams.JAMS()

    #Setando os parametros do arquivo de notacao
    jam.file_metadata.duration = duration
    ann = jams.Annotation(namespace='beat', time=0, duration=jam.file_metadata.duration)
    ann.annotation_metadata = jams.AnnotationMetadata(data_source=data_source)
    ann.annotation_metadata = jams.AnnotationMetadata(validation= "")
    ann.annotation_metadata.curator = jams.Curator(name=name, email=email)
    jam.annotations.append(ann)

    #Salvando o arquivo de notacao para cada audio
    jam_name = re.sub(r'\.wav', ".jams", audio_name)
    jam.save(jam_name)
```

```python
def audio_augmentation(audio_name, path_audio,fold_):

    files= glob.glob(path_backgound_noise + '/*.' + 'wav')

    #Captura o nome do audio sem extensa
    audio_name_wo_ext = re.sub(r'\.wav', "", audio_name)
      
    #Captura o caminho completo no arquivo de anotacao
    jam_path = re.sub(r'\.wav', ".jams", path_audio)
            
    #Captura o label da classe
    label = audio_name.split('-')[1]

    # A variacao de semitons e maior para as classes 1 e 6 
    if label == '1' or label == '6':
        semitones=[1,1.5,2,2.5,3,3.5]
    else:
        semitones=[1.5,2,2.5,3]

    jam_orig = muda.load_jam_audio(jam_path, path_audio) 

    #Gerando variacoes no tom
    ps = muda.deformers.PitchShift(n_semitones=semitones)
    output_name_ps_pattern = fold_+'/'+audio_name_wo_ext+ '-ps-'

    #O audio modificado e salvo em um novo .wav e as deformacoes aplicadas sao salvas no seu respectivo .jams
    for i, jam_out in enumerate(ps.transform(jam_orig)):
        muda.save(output_name_ps_pattern+str(semitones[i])+'.wav', output_name_ps_pattern+str(semitones[i])+'.jams',jam_out) 

    #Gerando deformacoes de ruido de fundo
    bg = muda.deformers.BackgroundNoise(n_samples=1, files=files, weight_min=0.1, weight_max=0.5)
    output_name_bg_pattern = fold_+'/'+audio_name_wo_ext+'-bg-'

    #O audio modificado e salvo em um novo .wav e as deformacoes aplicadas sao salvas no seu respectivo .jams
    for i, jam_out in enumerate(bg.transform(jam_orig)):
        bg_noise_name = files[i].split('/')[8]
        bg_noise_name_w_extension = re.sub(r'\.wav', "", bg_noise_name)
        print(output_name_bg_pattern+bg_noise_name_w_extension+'.wav')
        muda.save(output_name_bg_pattern+bg_noise_name_w_extension+'.wav', output_name_bg_pattern+bg_noise_name_w_extension+'.jams',jam_out)
```

```python
def run():
       
    #Capturando os nomes das pastas
    path_folders = glob.glob(path_database + '*')

    for fold_ in path_folders:
        
        print(fold_)
        
        #Listar o caminho completo dos audios
        path_audios = glob.glob(fold_ + '/*.' + 'wav') 

        for path_audio in path_audios:

            #Capturando o nome do arquivo de audio (wav) 
            match_obj = re.sub(fold_, "", path_audio)
            audio_name = re.sub(r'/', "", match_obj)

            #mudando o diretório para a pasta
            os.chdir(fold_+'/')
            
            #Gerando os jams
            jams_generation(audio_name)
            
            #Aumentando os dados
            audio_augmentation(audio_name,path_audio,fold_)
```

```python
run()
```
